## Walking Speed Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('processed_data.csv')
data['visit_date'] = pd.to_datetime(data['visit_date'])

#Scatter age - walking speed
plt.figure(figsize=(10, 6))
sns.lmplot(x="age", y="walking_speed", data=data, aspect=1.5)
plt.title("Age vs Walking Speed (Regression Line)")
plt.show()
```

```python
#Boxplot edu - walking speed
plt.figure(figsize=(10, 6))
sns.boxplot(x="education_level", y="walking_speed", data=data)
plt.title("Walking Speed by Education Level")
plt.show()
```

```python
#Lineplot edu - age
data['education_level'] = data['education_level'].astype('category')
sns.lineplot(x="age", y="walking_speed", hue="education_level", data=data, ci=None)
plt.title("Walking Speed: Education-Age Interaction")
plt.show()
```

## Cost Analysis

```python
#Barplot: insurance - cost
plt.figure(figsize=(10, 6))
sns.barplot(x="insurance_type", y="visit_cost", data=data, ci="sd")
plt.title("Mean Costs by Insurance Type")
plt.show()
```

```python
#Boxplot: insurance - cost
plt.figure(figsize=(10, 6))
sns.boxplot(x="insurance_type", y="visit_cost", data=data)
plt.title("Cost Distribution by Insurance Type")
plt.show()
```

## Advanced Analysis

```python
# Pair plot: Speed, cost, age
sns.pairplot(data, vars=["walking_speed", "visit_cost", "age"], hue="education_level")
plt.suptitle("Pair Plot of Key Variables", y=1.02)
plt.show()
```

```python
#Faceted plot: age - walking speed segmented by insurance and edu
g = sns.FacetGrid(data, col="education_level", row="insurance_type", margin_titles=True)
g.map_dataframe(sns.scatterplot, x="age", y="walking_speed")
g.set_axis_labels("Age", "Walking Speed")
plt.suptitle("Faceted Plot: Education and Insurance", y=1.02)
plt.show()
```
